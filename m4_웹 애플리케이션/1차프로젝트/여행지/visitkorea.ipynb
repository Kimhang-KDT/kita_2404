```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import requests
import pandas as pd
import re
from bs4 import BeautifulSoup
from datetime import datetime
import numpy as np
```

```python
# Chrome 옵션 설정
options = Options()

options.add_argument("--user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36")

# 웹 드라이버 초기화
driver = webdriver.Chrome(options=options)
driver.get('https://korean.visitkorea.or.kr/list/travelinfo.do?service=ms')
time.sleep(5)  # 페이지 로드 대기

data = []

# 페이지 ID 범위 설정
start_page_id = 1
end_page_id = 3925

for page_id in range(start_page_id, end_page_id + 1):
  try:
    # 페이지 이동을 위한 버튼 클릭
    driver.execute_script(f"document.querySelector('div.page_box > a[id=\"{page_id}\"]').click();")
    time.sleep(2)  # 페이지 로드 대기

    # 데이터 추출
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    lists = soup.select('div.area_txt.catchphrase')

    for list in lists:
      name = list.select_one('div.tit > a').text
      address = list.select_one('p').text
      spans = list.select('p.tag > span')
      tags = [tag.text for tag in spans]
      data.append([{'name': name, 'address': address, 'tags': tags}])

    print(f"Page {page_id} done")  # 각 페이지 완료 후 출력

  except Exception as e:
    print(f"Error on page {page_id}: {e}")
    break

# 추출된 데이터 출력
for dt in data:
  print(dt)

# 웹 드라이버 종료
driver.quit()
```
