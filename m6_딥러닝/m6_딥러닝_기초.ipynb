## 퍼셉트론
- 인공신경망의 기초가 되는 가장 간단한 형태의 알고리즘 중 하나. 입력(features)에 대해 가중치(weights)를 곱한 후, 이들의 합에 활성화 함수(activation function)를 적용하여 출력(output)을 생성. 퍼셉트론의 목표는 학습 데이터셋을 바탕으로 가중치를 조정하는 것이며, 이를 통해 주어진 입력 데이터를 두 개의 클래스 중 하나로 분류. 1957년에 프랭크 로젠블라트(Frank Rosenblatt)가 제안
- 인공지능(AI) 분야, 특히 신경망과 딥러닝 분야의 기본 구성 요소. 이는 원래 이진 분류 작업을 수행하도록 설계된 가장 간단한 형태의 신경망. 퍼셉트론과 그 한계, 그리고 오류 역전파 및 활성화 기능과 같은 신경망 설계의 후속 발전을 이해하는 것은 현대 AI 시스템의 개발과 기능을 이해하는 데 매우 중요.

인간의 뇌는 매우 복잡하고, 수많은 뉴런과 시냅스를 통해 이루어진 네트워크로, 이러한 신호 전달 과정이 순간적으로 일어난다.
- 뉴런은 외부 또는 내부로부터의 자극(예: 빛, 소리, 화학물질, 기계적 압력 등)을 수상돌기에서 수신
- 자극이 충분히 강하면, 뉴런은 액션 포텐셜(action potential)이라는 전기 신호를 생성
- 생성된 액션 포텐셜은 뉴런의 축삭(axon)을 따라 신속하게 전달
뉴런의 끝(축삭 말단)에 도달하면, 신경 전달물질(neurotransmitters)이 신경세포 간의 공간인 시냅스(synapse)로 방출
- 신경 전달물질은 시냅스를 건너 다음 뉴런의 수상돌기에 위치한 수용체(receptors)와 결합. 이 결합은 다음 뉴런에서 새로운 액션 포텐셜을 유발.

주요 이슈:

  - 이러한 한계는 다층 네트워크의 탐구와 비선형 문제를 해결할 수 있는 다층 퍼셉트론(MLP)의 개발로 이어졌다.
2. 오차 역전파
  - 더 심층적인 네트워크를 훈련하려면 가중치를 업데이트하는 더 효율적인 방법이 필요했다. 1980년대에 도입된 역전파 알고리즘은 다층 네트워크의 효율적인 훈련을 가능하게 했다. 이는 체인 규칙에 따라 각 가중치에 대한 손실 함수의 기울기를 계산하여 네트워크를 통해 출력 계층에서 입력 계층으로 오류를 효과적으로 "전파"하는 방식으로 작동한다.
  이 방법을 사용하면 네트워크 출력의 전체 오류를 최소화하는 방식으로 가중치를 조정할 수 있다.
3. 그라데이션 소멸 문제
  - 네트워크가 깊어짐에 따라 훈련 중에 레이어를 통해 다시 전파됨에 따라 기울기가 점점 더 작아지는 기울기 소멸 문제에 직면. 이로 인해 초기 레이어의 가중치를 업데이트하기가 어려워지고 딥 네트워크 훈련 속도가 느려지거나 심지어 중단되기도 한다.
  - 대체 활성화 함수(예: ReLU) 도입, 정규화 기술 및 특수 네트워크 아키텍처(예: 순환 네트워크용 LSTM)와 같은 혁신이 이 문제를 완화하는 데 도움이 되었다.
  - 활성화 함수의 선택은 네트워크가 복잡한 패턴을 학습하는 능력에 중요한 역할을 한다. 초기 네트워크는 간단한 계단 함수를 사용했지만 입력과 출력 간의 비선형 매핑을 학습할 수 있는 비선형 함수로 대체되었다.
  일반적인 활성화 함수에는 Sigmoid, tanh 및 ReLU(Rectified Linear Unit)가 포함. 각각에는 고유한 장점과 사용 사례가 있으며, ReLU와 그 변형(예: Leaky ReLU, ELU)은 Vanishing Gradient 문제를 완화하는 효율성과 효과로 인해 딥 러닝에서 특히 인기가 있다.

퍼셉트론은 오늘날 우리가 알고 있는 신경망과 AI의 토대를 마련했다. 그러나 초기 한계를 극복하려면 네트워크 아키텍처, 훈련 알고리즘 및 수학적 이해의 상당한 발전이 필요했다. 이미지 인식부터 자연어 처리까지 복잡한 작업을 처리할 수 있는 오늘날의 딥 러닝 시스템은 이러한 기본 개념을 기반으로 구축되었지만 해당 분야에서 수십 년간의 연구 개발 덕분에 훨씬 더 정교하게 진화했다.

```python
import matplotlib.pyplot as plt
import numpy as np

# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# XOR 입력과 목표 값
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])

# 가중치 초기화
np.random.seed(42)
w_h = np.random.randn(2, 2)  # 은닉층 가중치
b_h = np.random.randn(2)     # 은닉층 편향
w_o = np.random.randn(2)     # 출력층 가중치
b_o = np.random.randn()      # 출력층 편향

# 순방향 전달 예시
def forward(X):
    h = sigmoid(np.dot(X, w_h) + b_h)
    o = sigmoid(np.dot(h, w_o) + b_o)
    return h, o

# 은닉층 및 출력 계산
h, o = forward(X)

# 결과 시각화
plt.figure(figsize=(10, 5))

# 입력 데이터
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')
plt.title('XOR Input Data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.grid(True)

# 은닉층 출력
plt.subplot(1, 2, 2)
plt.scatter(h[:, 0], h[:, 1], c=y, cmap='bwr')
plt.title('Hidden Layer Output')
plt.xlabel('h1')
plt.ylabel('h2')
plt.grid(True)

plt.show()
```

#### 다층 퍼셉트론으로 XOR 문제 해결하기
XOR 문제는 선형적으로 분리되지 않는 문제로, 단층 퍼셉트론(단일층 신경망)으로는 해결할 수 없습니다. 이를 해결하기 위해 다층 퍼셉트론(MLP, Multi-Layer Perceptron)을 사용해야 합니다. 다층 퍼셉트론은 최소 한 개 이상의 은닉층(hidden layer)을 포함하여 입력과 출력을 비선형적으로 변환할 수 있습니다.

```python
import numpy as np

# 가중치와 바이어스
w11 = np.array([-2, -2])
w12 = np.array([2, 2])
w2 = np.array([1, 1])
b1 = 3
b2 = -1
b3 = -1

# 퍼셉트론 함수 정의
# 활성화 함수는 계단 함수 적용(특정 임계값을 기준으로 출력을 0 or 1로 결정)
def MLP(x, w, b):
    y = np.sum(w * x) + b
    if y <= 0:
        return 0
    else:
        return 1

# NAND 게이트
def NAND(x1, x2):
    return MLP(np.array([x1, x2]), w11, b1)

# OR 게이트
def OR(x1, x2):
    return MLP(np.array([x1, x2]), w12, b2)

# AND 게이트
def AND(x1, x2):
    return MLP(np.array([x1, x2]), w2, b3)

# XOR 게이트: NAND, OR, AND 게이트를 적절히 조합하여 XOR의 비선형성을 구현
def XOR(x1, x2):
    return AND(NAND(x1, x2), OR(x1, x2))

# x1, x2 값을 번갈아 대입해 가며 최종 값 출력
for x in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = XOR(x[0], x[1])
    print("입력 값: " + str(x) + " 출력 값: " + str(y))
```

미분의 필요성

- 최적화 방향 제공: 미분을 통해 계산된 기울기는 손실을 최소화하기 위해 파라미터를 업데이트하는 데 필요한 방향과 크기를 제공.
- 안정적인 학습 과정: 적절한 미분 계산을 통해 학습률(learning rate)과 같은 하이퍼파라미터를 더 잘 조절할 수 있으며, 이는 학습 과정을 안정적으로 만들고 오버슈팅(overshooting) 또는 느린 수렴과 같은 문제를 방지할 수 있다.

기울기 업데이트와 미분은 모델을 학습시키고 최적화하는 데 있어 상호 보완적인 역할을 한다. 미분을 통해 계산된 기울기를 사용하여 가중치를 업데이트하고, 이 과정을 반복함으로써, 모델은 점차 최적의 파라미터로 수렴하게 된다. 이러한 과정은 다양한 기계 학습 모델과 알고리즘에서 핵심적인 메커니즘으로 작용.


```python
# 필요한 라이브러리 임포트
import matplotlib.pyplot as plt
import numpy as np

# 폰트 및 마이너스 기호 설정
plt.rc('font', size=12)  # 로컬 환경에 따라 'family' 설정 필요
plt.rcParams['axes.unicode_minus'] = False

# 주어진 함수와 도함수 정의
def f(x):
    """주어진 함수 f(x) = x^2"""
    return x**2

def f_prime(x):
    """함수 f의 도함수 f'(x) = 2x"""
    return 2*x

def tangent_line(f_prime, x, x1, y1):
    """점 (x1, y1)에서 함수 f의 접선의 방정식 y = mx + b에서 x = x-x1, b = y1 (x=>x1일 때 y=>y1)"""
    return f_prime(x1) * (x - x1) + y1

def plot_function_and_tangents(x, x_points):
    """함수와 접선들을 그리는 함수, x_points에서의 접선 계산 및 그리기 로직 포함"""
    fig = plt.figure(figsize=(6, 6))
    plt.scatter(x_points, [f(x1) for x1 in x_points], c='k')  # 점들 표시
    plt.plot(x, f(x), 'k', label='$f(x) = x^2$')  # 함수 f(x) 그래프

    # 각 x_points에서의 접선 계산 및 그리기
    for x1 in x_points:
        y1 = f(x1)
        tangent = tangent_line(f_prime, x, x1, y1)  # 접선 계산
        plt.plot(x, tangent, label=f'Tangent at x={x1}')  # 접선 그리기

    plt.axhline(0, lw=2, color='k')  # x축
    plt.axvline(0, lw=2, color='k')  # y축
    plt.ylim(-1, 9)
    plt.xlim(-3, 3)
    plt.xlabel('x', fontsize=14)
    plt.ylabel('y', fontsize=14)
    plt.grid(lw=1)
    plt.legend()
    plt.show()

# x 좌표 배열과 접선을 그릴 x의 점들 정의
x = np.linspace(-3, 3, 100)
x_points = [1, 2]  # 접선을 추가할 x의 점들
plot_function_and_tangents(x, x_points)
```

## 접선의 방정식

- 함수 f(x)에서 x의 값을 dx만큼 변화시켰을 때 f(x)의 변화량 (f(x+dx)-f(x))는 f'(x)dx와 같다.
- 미분 y=f(x)라는 그래프상의 점 (a,f(a))가 있을 때 그 점에서 그은 접선의 기울기는 f'(a)이며 a에서의 순간 변화율을 의미
- 적분 : 도함수 f'(x)의 하방 면적은 원 함수 f(x)의 y의 값이다 ex) f(x) = x ** 2

Q. 주어진 함수 $ f(x, y) = x^2 + y^2 $에 대해 x에 대한 편미분과 y에 대한 편미분을 구하세요.

```python
x, y = sympy.symbols('x y')
f = x ** 2 + y ** 2

# x에 대한 편미분
partial_x = sympy.diff(f, x)

# y에 대한 편미분
partial_y = sympy.diff(f, y)

partial_x, partial_y
```

주어진 함수에 대해서,

$$
f(x, y) = x^2 + y^2
$$

- \(x\)에 대한 편미분을 구하면,

$$
\frac{\partial f}{\partial x} = \frac{\partial}{\partial x}(x^2 + y^2) = 2x
$$

- \(y\)에 대한 편미분을 구하면,

$$
\frac{\partial f}{\partial y} = \frac{\partial}{\partial y}(x^2 + y^2) = 2y
$$

### 경사하강법(Gradient Descent)
- 머신러닝과 딥러닝에서 비용 함수(Cost Function) 또는 손실 함수(Loss Function)의 값을 최소화하기 위해 사용되는 최적화 알고리즘. 이 방법은 주어진 함수의 최소값을 찾기 위해 반복적으로 파라미터를 조정해 나가는 방식으로 작동.
- 경사하강법의 기본 아이디어는 함수의 기울기(경사)를 계산하고, 그 기울기가 감소하는 방향으로 파라미터를 업데이트하는 것이다. 기울기는 함수의 현재 위치에서 가장 가파르게 증가하는 방향을 나타내므로, 반대 방향으로 이동하면 함수의 최소값을 찾을 수 있다.

가중치 a에 대한 편미분

오차 함수 E를 가중치 a에 대해 편미분하여 가중치 a의 변화가 오차에 미치는 영향을 구합니다.

$\frac{\partial E}{\partial a} = \frac{2}{n} \sum_{i=1}^{n} -x_i (y_i - (a x_i + b))$

<br>
가중치 a 업데이트

가중치 a를 학습률 lr과 편미분 값(기울기)을 사용하여 업데이트합니다.

$a = a - lr \cdot \frac{\partial E}{\partial a}$

```python
import numpy as np
import matplotlib.pyplot as plt

# 공부시간 X와 성적 Y의 넘파이 배열을 만듭니다.
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

# 기울기 a와 절편 b의 값을 초기화 합니다.
a = 0
b = 0

# 학습률 : 경사 하강법에서 매 반복마다 가중치(파라미터)를 얼마나 업데이트할지를 결정
lr = 0.03

# 몇 번 반복할지를 설정합니다.
epochs = 2001

# x값이 총 몇 개인지 셉니다.
n = len(x)

# 경사하강법을 시작합니다.
for i in range(epochs): # epoch 수 만큼 반복

  y_pred = a * x + b # 예측 값을 구하는 식
  error = y - y_pred # 실제 값과 비교한 오차를 error로 놓습니다.

  a_diff = (2/n) * sum(-x * (error)) # 오차 함수를 a로 편미분한 값입니다.
  b_diff = (2/n) * sum(-(error)) # 오차 함수를 b로 편미분한 값입니다.

  a = a - lr * a_diff # 학습률과 편미분 값을 사용해 기존의 a 값을 업데이트 합니다.
  b = b - lr * b_diff # 학습률을 곱해 기존의 b 값을 업데이트 합니다.

  if i % 100 == 0:
    print('epoch=%.f, 기울기=%.04f, 절편=%.04f' % (i, a, b))
```

```python
# 앞서 구한 최종 a값을 기울기, b값을 y절편에 대입하여 그래프를 그립니다.
y_pred = a * x + b
print(a, b)

# 그래프 출력
plt.scatter(x, y)
plt.plot(x, y_pred, 'r')
plt.show()
```

### 오차 역전파(Backpropagation)
신경망을 훈련시키기 위해 사용되는 핵심 알고리즘 중 하나. 이 방법은 신경망의 출력(output)과 실제 값(target) 사이의 오차를 계산하고, 이 오차를 신경망을 거슬러 올라가며 각 뉴런의 가중치를 조정하는 방식으로 작동.<BR>
오차역전파는 기본적으로 경사 하강법(Gradient Descent)의 원리를 사용하여 신경망의 가중치를 최적화.<BR>

오차 역전파는 두 단계로 이루어진다: 순전파(Forward Pass)와 역전파(Backward Pass).
- 순전파(Forward Pass):
  - 입력 데이터는 신경망의 각 층을 순차적으로 통과하면서, 각 뉴런의 가중치와 활성화 함수를 통해 처리.
  - 마지막 층에서의 출력값은 신경망의 예측값이 된다.
- 역전파(Backward Pass):
  - 신경망의 예측값과 실제 값 사이의 오차를 계산.
  - 이 오차를 바탕으로, 오차 함수의 기울기(미분값)를 계산하고, 이 기울기 정보를 신경망을 거슬러 올라가며 전달.
  - 각 층의 가중치는 계산된 기울기를 사용하여 조정. 이 과정은 신경망의 예측 오차를 줄이는 방향으로 가중치를 업데이트.

중요성
  - 오차역전파는 심층 신경망(Deep Neural Networks)의 훈련에 필수적인 기법으로 이 알고리즘을 통해 복잡한 신경망 구조에서도 효율적으로 가중치를 조정할 수 있다.
  - 이 방법은 신경망의 각 층에서 발생하는 오차의 기여도를 정확히 계산하고, 이를 통해 신경망의 학습이 어떻게 이루어지는지 이해할 수 있게 한다.

오차역전파를 통한 학습 과정에서 적절한 학습률과 초기 가중치 선택이 중요. 잘못된 선택은 학습이 느려지거나, 지역 최소값에 빠지는 등의 문제를 일으킬 수 있다.

신경망의 깊이가 깊어질수록 기울기 소실(Vanishing Gradient) 또는 기울기 폭발(Exploding Gradient) 문제가 발생할 수 있으며, 이를 해결하기 위한 다양한 기법들(예: 가중치 초기화 방법, 정규화 기법, 배치 정규화 등)이 제안되고 있다.

### 기울기 소실(Vanishing Gradient)
- 심층 신경망(Deep Neural Networks)을 훈련시킬 때 발생할 수 있는 현상으로, 신경망의 깊이가 깊어질수록 오차역전파 과정에서 계산된 기울기(오차의 미분값)가 점점 작아져, 입력층에 가까운 층의 가중치가 거의 또는 전혀 업데이트되지 않는 문제를 의미.
- 이 현상은 특히 활성화 함수로 시그모이드(Sigmoid)나 하이퍼볼릭 탄젠트(Hyperbolic Tangent)와 같은 함수를 사용할 때 두드러진다.
- 기울기 소실 문제의 주된 원인은 신경망을 거슬러 올라가며 기울기를 곱해 나가는 과정에서, 기울기 값이 점점 줄어들어 0에 가까워지기 때문입니다. 특히, 활성화 함수의 도함수 값이 0과 1 사이에 있을 때(예: 시그모이드 함수), 신경망이 깊어질수록 기울기가 점점 더 작아지게 된다.
- 기울기 소실 문제가 발생하면, 신경망의 초기 층의 가중치가 충분히 업데이트되지 않아, 모델의 학습이 제대로 이루어지지 않는다. 이는 신경망이 복잡한 패턴이나 특징을 학습하는 데 어려움을 겪게 만들며, 결국 모델의 성능 저하로 이어진다.
- 기울기 소실 문제를 해결하기 위한 몇 가지 방안이 제안되었다:
  - ReLU 활성화 함수 사용: ReLU(Rectified Linear Unit)와 같은 활성화 함수는 양의 입력에 대해 기울기가 1이 되므로, 기울기 소실 문제를 어느 정도 완화할 수 있다.
  - 가중치 초기화 방법 개선: 효과적인 가중치 초기화 방법을 사용하여 신경망의 초기 가중치를 적절히 설정.
  - 정규화 기법 사용: 배치 정규화(Batch Normalization) 등의 정규화 기법을 사용하여 각 층의 입력 분포를 안정화시키고, 기울기 소실 문제를 완화.
  - 잔차 연결(Residual Connections): 잔차 네트워크(ResNet)와 같이 잔차 연결을 통해 입력을 출력에 직접 더함으로써, 신경망이 깊어져도 기울기가 잘 전파될 수 있도록 한다.

### 2. 차이점: 네트워크 구성
딥러닝 모델들의 주요 차이점은 **네트워크 구성**(구조, 아키텍처)에서 발생합니다. 네트워크의 구성은 모델이 특정 문제를 얼마나 잘 해결할 수 있는지에 큰 영향을 미치며, 여기서 다양한 모델들이 등장하게 됩니다. 아래에 몇 가지 대표적인 네트워크 유형을 설명하겠습니다.

#### (1) **다층 퍼셉트론 (MLP, Multi-Layer Perceptron)**
- **구성**: Fully Connected 레이어들로 구성된 가장 기본적인 신경망.
- **적용**: 주로 구조가 간단한 데이터(예: tabular data)에 사용됩니다.
- **특징**: 입력과 출력 사이에 여러 개의 전결합(fully connected) 레이어가 있으며, 모든 뉴런이 서로 연결되어 있습니다.

#### (2) **컨볼루션 신경망 (CNN, Convolutional Neural Network)**
- **구성**: 컨볼루션 레이어와 풀링 레이어를 포함하여 입력 데이터를 처리하는 구조.
- **적용**: 이미지 처리, 영상 분류, 물체 탐지 등에서 매우 효과적입니다.
- **특징**: 이미지와 같은 고차원 데이터를 처리할 때 적합한 아키텍처로, 입력 데이터의 공간적 구조를 보존하며 학습합니다. 컨볼루션 레이어는 특징 맵(feature map)을 추출하고, 풀링 레이어는 차원을 축소하여 중요한 정보를 남깁니다.

#### (3) **순환 신경망 (RNN, Recurrent Neural Network)**
- **구성**: 시퀀스 데이터를 처리하는데 적합한 네트워크로, 이전 상태의 출력을 다음 상태의 입력으로 사용.
- **적용**: 시계열 데이터, 자연어 처리(NLP), 음성 인식 등에 사용됩니다.
- **특징**: 시퀀스 간의 종속성을 학습할 수 있으며, 시간적 정보를 고려하는 모델입니다. LSTM(Long Short-Term Memory)과 GRU(Gated Recurrent Unit)와 같은 변형이 널리 사용됩니다.

#### (4) **변환기 (Transformer)**
- **구성**: 셀프 어텐션(self-attention) 메커니즘을 사용하는 아키텍처로, 병렬 처리가 가능하여 효율적입니다.
- **적용**: 자연어 처리(NLP), 특히 번역, 요약 등의 작업에 주로 사용됩니다. 최근에는 이미지 처리와도 결합되어 사용됩니다.
- **특징**: RNN과 달리 시퀀스를 순차적으로 처리하지 않고, 셀프 어텐션 메커니즘을 사용해 문맥 상의 단어 관계를 학습합니다. GPT, BERT 같은 모델이 이 구조를 따릅니다.

#### (5) **생성적 적대 신경망 (GAN, Generative Adversarial Network)**
- **구성**: 두 개의 신경망(생성자(generator)와 판별자(discriminator))이 경쟁하는 구조.
- **적용**: 이미지 생성, 데이터 증강 등에 사용됩니다.
- **특징**: 생성자는 실제 데이터를 모방하여 새로운 데이터를 생성하고, 판별자는 이 데이터가 진짜인지 가짜인지 판별합니다. 이 두 네트워크는 경쟁적으로 학습하며, 결국 생성자는 매우 사실적인 데이터를 생성할 수 있게 됩니다.

### 3. 네트워크 구성의 중요성
네트워크의 구성에 따라 딥러닝 모델의 성능과 그 모델이 처리할 수 있는 데이터의 유형이 결정됩니다. 예를 들어:

- CNN은 **공간적 데이터**(이미지나 영상)에서 매우 효과적인 반면, MLP는 이러한 데이터를 처리하는 데 한계가 있습니다.
- RNN과 트랜스포머는 **시퀀스 데이터**(문장, 음성 등)에서 매우 뛰어난 성능을 보이지만, CNN처럼 2D 이미지를 처리하는 데는 적합하지 않습니다.
- GAN은 **데이터 생성**에 특화되어 있으며, 이를 통해 가짜 데이터를 생성하는 데 매우 유용합니다.

### 결론:
딥러닝 모델은 기본적으로 같은 신경망 알고리즘(예: 경사 하강법, 역전파, 손실 함수 최소화)을 사용합니다. 하지만 그 모델의 **구성(네트워크 아키텍처)**에 따라 성능과 적용 가능성이 크게 달라집니다. 각 아키텍처는 문제에 특화된 특징을 가지고 있으며, 그에 맞는 구조를 설계하는 것이 중요합니다.
