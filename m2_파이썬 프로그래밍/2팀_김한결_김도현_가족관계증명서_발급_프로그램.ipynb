```python
-가족관계 증명서 발급 및 출력 프로그램-

1. class FamilyInfo(User) : 나의 가족의 정보가 담긴 dict family_info 생성 # ssn의 뒷자리는 앞자리만 빼고 *로 표시
  - 'User'를 기준으로 부, 모, 배우자, 자녀 정보가 입력됨 # 출력시 부-모-배우자-자녀 순으로 표시
  - family_info{
    { name:'부', ssn:주민번호, birth:생년월일, sex:남, rel:부 }, { name:'모', ssn:주민번호, birth:생년월일, sex:여, rel:모 },
    { name:'배우자', ssn:주민번호, birth:생년월일, sex:남, rel:배우자 }, { name:'자녀1', ssn:주민번호, birth:생년월일, sex:여, rel:자녀 },
    { name:'자녀2', ssn:주민번호, birth:생년월일, sex:남, rel:자녀 } }
  - 본인의 정보 dict(user_info) 생성 : class UserInfo
  - user_info { name:'본인', address:주소, birth:생년월일, sex:여, phone:핸드폰번호, ssn:주민번호 }

2. class User(UserInfo) 만든다. User는 UserInfo를 상속받는다.
  - self.auth = False # 디폴트
  - home() : print('가족관계증명서 발급 페이지입니다.') # user.auth 검사
3. class Auth(User) : 본인인증 진행
  - Authenticate() : User로부터 이름, 핸드폰번호, 생년월일을 받아 user_info와 맞는지 판별 후 True 일시 authnum(인증번호) 반환
4. User.auth가 True인지 False인지 판별하여 False이면 본인인증 진행
  - print('본인인증이 필요한 페이지입니다. 본인인증을 시작합니다.')
  - input('이름을 입력해주십시오(홍길동)')
  - input('핸드폰 번호를 입력해주십시오(01012345678)')
  - input('(생년월일 8자리를 입력해주십시오(20000101))')
  - print('입력하신 번호로 인증메시지를 보냈습니다.')
  - print(인증번호) # 인증 번호 4자리 랜덤 생성 후 출력
  - input('인증번호를 입력해주세요')
5. 인증번호 True False 판별 후 리턴 # User.auth = True
6. home() 다시 실행 user.auth 검사
7. if user.auth == True : input('가족관계증명서를 발급 하시겠습니까?')
  - 발급 입력 : familyinfo의 데이터를 사용하려면 user.auth가 True여야함
8. FamilyInfo.family_info를 불러와 데이터 세팅 후 출력

# 출력
등록 기준지 |
   구분   |  성명  |   출생연월일     |    주민번호    | 성별
   본인     홍길순   1980년 02월 10일   800210-2123456    여

가족사항
   구분   |  성명  |   출생연월일     |    주민번호    | 성별
    부      홍철수   1945년 04월 12일   450412-1243475    남
    모      이영미   1943년 07월 04일   430704-2652535    여
  배우자    박배우   1982년 11월 16일   821116-1336412    남
   자녀     박자순   2005년 12월 09일   051209-4566442    여
   자녀     박자돌   2007년 08월 03일   070803-3634646    남
```

```python
import random
import time
import re

#class UserInfo:
#  def __init__(self):
#    self.user_info = { 'name':'홍길순', 'address':'서울 강남구 테헤란로 146', 'birth':'19800210', 'sex':'여', 'phone':'01012345678', 'ssn':'800210-2123456' }

class FamilyInfo:
  def __init__(self):
    # familycode : 1 하위 dict로 만들어서 다른 더미 데이터와 함께 섞어 familycode가 같은 사람들끼리 묶는 방법도 있음
    # 이렇게 할때에는 rel표현할때 본인 코드를 0으로 가정하여 부는 1 모는 2 배우자는 0 자녀는 -1 이런식으로 묶는 방법도 있음
    self.family_info = [
      [{ 'name':'홍철수', 'ssn':'450412-1243475', 'birth':'19450412', 'sex':'남', 'rel':'부' }],
      [{ 'name':'이영미', 'ssn':'430704-2652535', 'birth':'19430704', 'sex':'여', 'rel':'모' }],
      [{ 'name':'박배우', 'ssn':'821116-1336412', 'birth':'19821116', 'sex':'남', 'rel':'배우자' }],
      [{ 'name':'박자순', 'ssn':'051209-4566442', 'birth':'20051209', 'sex':'여', 'rel':'자녀' }],
      [{ 'name':'박자돌', 'ssn':'070803-3634646', 'birth':'20070803', 'sex':'남', 'rel':'자녀' }]]
    self.user_info = { 'name':'홍길순', 'address':'서울 강남구 테헤란로 146', 'birth':'19800210', 'sex':'여', 'phone':'01012345678', 'ssn':'800210-2123456' }

class User(FamilyInfo):
  def __init__(self):
    super().__init__()
    self.auth = False

class Auth(User):
  def __init__(self):
    super().__init__()

  def authenticate(self):
    if self.auth:
      return True

    print('본인인증이 필요한 페이지입니다. 본인인증을 시작합니다.')

    name, phone, birth = input_user_info()
    info = self.user_info

    if (name, phone, birth) == (info['name'], info['phone'], info['birth']):
      while True:
        print('입력하신 번호로 인증번호를 보냈습니다.')
        auth_num = ''.join(str(i) for i in random.sample(range(0,10), 4)) # 0~9까지의 랜덤 정수 중 4개
        time.sleep(1)
        print(f"인증번호 [ {auth_num} ]") # 인증 번호 4자리 랜덤 생성 후 출력
        num = input('인증번호를 입력해주세요 : ')
        if num != auth_num:
          print('잘못 입력하셨습니다. 다시 시도해주세요')
          continue

        return True
    else:
      print('해당 사용자는 정보가 존재하지 않습니다.')
      return False

def input_user_info():
  while True:
    name  = input('이름을 입력해주십시오(홍길순) : ')
    phone = input('핸드폰 번호를 입력해주십시오(01012345678) : ')
    birth = input('(생년월일 8자리를 입력해주십시오(19800210)) : ')
    if not name.isalpha() or not re.match(r'^010\d{8}$', phone) or not  re.match(r'\d{8}', birth):
      print('잘못 입력하셨습니다. 다시 입력해주십시오')
      continue
    else:
      return name, phone, birth

def home():
  print('가족관계증명서 발급 페이지입니다.')

def main():
  auth = Auth()
  home() # 홈 멘트 출력

  while True:
    if auth.authenticate():
      auth.auth = True
      break
    else:
      continue

  print(auth.auth) # True
  print(f'유저 정보가 확인되었습니다.\n환영합니다 {auth.user_info["name"]}님!')
  if input('가족관계증명서를 발급 하시겠습니까? (Y|N) : ').upper() == 'Y':
    print('가족관계증명서를 발급 중입니다...')
    time.sleep(3)
    for mem in auth.family_info[0]:
      print(mem)

if __name__ == '__main__':
  main()
```
