Q. 한 학급의 학생 10명의 수학 시험 점수는 다음과 같습니다:

[85, 72, 90, 68, 88, 76, 95, 89, 77, 84].

학급의 평균 점수와 표준편차를 계산하여 전체 성적 수준과 변동성을 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
arr = np.array([85, 72, 90, 68, 88, 76, 95, 89, 77, 84])
df = pd.DataFrame(arr)
print("평균점수 : ",df.mean(), "\n")
print("표준편차 : ",df.std())

# 평균은 학급의 전체 성적 수준을 나타내고, 표준편차는 성적의 변동성을 나타냅니다.
if df.std() > 10:
  print("성적 변동성이 큽니다.")
else:
  print("성적 변동성이 작습니다.")

# 인사이트
# 학급의 평균 점수는 82.4점. 100점 만점 기준으로 10등급으로 나눈다고 가정
# 1등과 2등 학생의 점수 차는 5점
# 평균 점수 이상의 점수를 가진 학생의 수는 6명, 70점 대의 학생 수는 3명, 60점대는 1명이다.
# 1등 학생과 꼴등 학생의 점수 차는 27점으로 표준편차의 3.1배 차이가 난다.
# 90점 이상의 학생들의 점수 차는 5점, 80점대 학생들의 표준편차는 1.6점, 70점대는 2.5점
# 학급의 성적수준이 비교적 높은 편이며, 성적의 변동성이 낮아 고르게 분포되어 있음을 알 수 있습니다.
```

Q. 어느 도시에 지난 1년간 매달 내린 강수량 데이터는 다음과 같습니다:

[85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95].

이 데이터를 이용해 중간값과 1사분위수, 3사분위수를 계산하여 강수량의 분포를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
prec = np.array([85, 90, 78, 110, 88, 76, 120, 89, 77, 91, 100, 95])

df = pd.DataFrame({
  'pref':prec
}, index=range(1,13))
df.index.name = 'months'

med = df.median()
q1 = df.quantile(0.25) # 1사분위수
q3 = df.quantile(0.75) # 3사분위수

#print(f"중간값 : {med}\n")
#print(f"1사분위수 : {q1}\n")
#print(f"3사분위수 : {q3}\n")

# 강수량의 분포를 파악
df.plot(kind='bar', color='skyblue')
plt.title('Precipitation Distribution Map')
plt.xlabel('months')
plt.ylabel('prec')
plt.xticks(rotation=0) # label 표기 회전율
plt.legend()
plt.show()

# 중간값은 전체 강수량 분포의 중심을 나타내고, 사분위수는 데이터의 분포 범위를 나타낸다.
# 강수량의 변동성은 q3-q1 (mm) : 13mm으로, 변동성이 비교적 작음을 알 수 있다.
# 중간값은 89.5mm로 일반적인 강수량 수준임을 나타낸다.

# 인사이트
# 7월과 4월의 강수량이 1년 중 가장 높은 것으로 확인
# 6월의 강수량이 가장 낮은 것으로 확인
# 4월부터 6월까지는 점차 강수량이 낮아지다 6월에 최저치를 찍고
# 7월에 급등 후 다시 낮아지다 9월에 최저 찍은 후 10월부터 12월까지 높은 강수량을 보임
```

Q. 한 회사의 직원 연봉 데이터가 다음과 같습니다:

[45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000].

이 데이터를 이용하여 연봉의 분포와 이상치를 시각화하는 상자 그림을 그리고, 연봉 분포에 대한 인사이트를 도출하세요.

```python
salary = np.array([45000, 50000, 52000, 47000, 48000, 53000, 60000, 62000, 55000, 58000])

plt.boxplot(salary)
plt.title('Box Plot of Employee Salary')
plt.ylabel('Salary')
plt.show()

# 인사이트
m_salary = salary.mean()
s_salary = salary.std()
print(f"평균 연봉 : {m_salary}")
print(f"표준편차 : {s_salary}")

q1_salary = np.quantile(salary, 0.25)
q3_salary = np.quantile(salary, 0.75)
iqr_salary = q3_salary - q1_salary

print(f"Q1: {q1_salary}")
print(f"Q3: {q3_salary}")
print(f"IQR: {iqr_salary}")

# 이상치 확인
lower_bound = q1_salary - 1.5 * iqr_salary # q1보다 1.5배 작으면 이상치
upper_bound = q3_salary + 1.5 * iqr_salary # q3보다 1.5배 크면 이상치

outliers = salary[(salary < lower_bound) | (salary > upper_bound)] # 둘 중 하나라도 속한다면 이상치
outliers

# 평균 연봉은 53000, 표준편차는 5422로 연봉의 변동성은 크지 않습니다.
# 연봉이 너무 낮거나 너무 높은 이상치는 존재하지 않습니다.
# q3-q1의 편차는 8750로 연봉이 고루 분포되어있음을 확인할 수 있습니다.
```

Q. 한 마케팅 캠페인에서 투자한 광고비와 발생한 매출 데이터가 각각 다음과 같습니다:

광고비 [1, 2, 3, 4, 5] (단위: 천 달러),

매출 [10, 15, 20, 25, 30] (단위: 천 달러).

광고비와 매출 간의 관계를 시각화하기 위해 산점도를 그리고, 두 변수 간의 관계에 대한 인사이트를 도출하세요.

```python
ad_cost = [1,2,3,4,5]
sales = [10,15,20,25,30]

df = pd.DataFrame({
  'ad_cost':ad_cost,
  'sales':sales
})
df.index.name = 'index'

# 광고비와 매출의 상관계수
corr = df.corr()
print(corr)

if corr > 0.7:
  print("광고비와 매출에 대한 관계가 강합니다.")
else:
  print("광고비와 매출 관계가 약합니다.")

# 광고비와 매출 산점도
plt.scatter(ad_cost, sales, color='skyblue')
plt.title('Ad Cost vs. Sales')
plt.xlabel('Ad Cost')
plt.ylabel('Sales')
plt.show()
```

Q. 한 식당의 월별 홍보 비용과 월별 방문자 수 데이터가 각각 다음과 같습니다:

홍보 비용 [500, 600, 700, 800, 900] (단위: 달러),

방문자 수 [300, 400, 500, 600, 700] (단위: 명).

홍보 비용과 방문자 수 간의 상관 계수를 계산하여 관계를 파악하세요. 이 데이터에서 어떤 인사이트를 얻을 수 있습니까?

```python
promotion_cost = [500, 600, 700, 800, 900]
visitors = [300, 400, 500, 600, 700]

df = pd.DataFrame({
  'promotion_cost': promotion_cost,
  'visitors': visitors
})

corr = df.corr()

if corr > 0.7:
  print("홍보비와 방문자 수에 강한 양의 상관관계가 있습니다.")
else:
  print("홍보비와 방문자 수에 상관관계가 약하거나 없습니다.")

# 증가하는 홍보비에 비례하여 방문하는 방문자의 수 또한 증가
# 홍보비 100달러 당 방문자 수 100명 증가
# 방문자 수 한명에 홍보비 1달러 소모

plt.scatter(df['promotion_cost'], df['visitors'])
plt.title('Promotion Cost vs. Visitors')
plt.xlabel('Promotion Cost')
plt.ylabel('Visitors')
plt.show()
```

Q. 한 지역의 주택 가격 데이터가 다음과 같습니다:

[250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000].

주택 가격의 분포를 시각화하기 위해 분포 그래프를 그리고, 주택 가격 분포에 대한 인사이트를 도출하세요.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

homes = np.array([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

df = pd.DataFrame({
  'homes':homes
})

# 주택 가격 중간값, 1사분위값 3사분위값, 표준편차
print(homes.mean())
print(homes.std())
print(np.quantile(homes, 0.25))
print(np.quantile(homes, 0.75))

# Bar Plot
df.plot(kind='bar', color='skyblue')
plt.title('Distribution of Home Prices')
plt.xlabel('Homes')
plt.ylabel('Prices')
plt.show()
```

```python
homes = np.array([250000, 300000, 350000, 400000, 450000, 500000, 550000, 600000, 650000, 700000])

# 밀도 그래프 그리기
sns.kdeplot(homes, fill=True)
plt.title('Distribution of Home Prices')
plt.xlabel('Prices')
plt.ylabel('Density')
plt.show()

# 인사이트
mean = homes.mean()
std = homes.std()
print(f"평균: {mean}")
print(f"표준편차: {std}")
med = np.median(homes)
print(f"중간값: {med}")
print(np.quantile(homes, 0.25))
print(np.quantile(homes, 0.75))

if mean > med:
  print("주택 가격 분포가 오른쪽으로 치우쳐 있음을 의미합니다.")
elif mean < med:
  print("주택 가격 분포가 왼쪽으로 치우쳐 있음을 의미합니다.")
else:
  print("주택 가격이 고르게 분포되어 있음을 의미합니다.")

# 평균과 중간값이 똑같다는것을 알 수 있습니다. 이로인해 주택 가격이 고르게 분포되어 있음을 알 수 있습니다.
```

Q. 한 교육기관의 시험 점수 데이터가 다음과 같습니다:

[55, 60, 65, 70, 75, 80, 85, 90, 95, 100].

시험 점수의 도수 분포표를 작성하고, 이를 통해 성적 분포에 대한 인사이트를 도출하세요.

```python
scores = np.array([55, 60, 65, 70, 75, 80, 85, 90, 95, 100])

# 시험 점수의 도수 분포
df = pd.DataFrame({
  'scores':scores
})

bins = [0, 60, 70, 80, 90, 101]
labels = ['F', 'D', 'C', 'B', 'A']
df['freq'] = pd.cut(df.scores, bins=bins, labels=labels, right=False)

# 학생 성적 freq별 Bar Plot
df.freq.value_counts().plot(kind='bar', color='skyblue')
plt.title('Frequency of Scores')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()

```

```python
# 시험 점수 데이터
scores = np.array([55, 60, 65, 70, 75, 80, 85, 90, 95, 100])

# 도수 분포표 생성
(unique, counts) = np.unique(scores, return_counts=True)
freq_table = np.asarray((unique, counts)).T

# 도수 분포표 출력
print("도수 분포표:")
print(freq_table)

# 인사이트
print("시험 점수의 도수 분포표를 통해 성적이 고르게 분포되어 있음을 확인할 수 있습니다.")
```

Q. 한 회사의 연구개발(R&D) 투자 비용과 연간 수익 데이터가 각각 다음과 같습니다:

투자 비용 [100, 200, 300, 400, 500] (단위: 천 달러),

연간 수익 [200, 400, 600, 800, 1000] (단위: 천 달러).

추세선이 있는 산점도를 그려 투자 비용과 연간 수익 간의 관계를 시각화하고, 인사이트를 도출하세요.

```python
inv_cost = [100, 200, 300, 400, 500]
ann_prof = [200, 400, 600, 800, 1000]

df = pd.DataFrame({
  'inv_cost':inv_cost,
  'ann_prof':ann_prof
})

# 추세선을 포함한 산점도
# 투자 비용과 연간 수익 간의 관계 시각화
plt.scatter(inv_cost, ann_prof, color='skyblue')
sns.regplot(x='inv_cost', y='ann_prof', data=df)
plt.title('Investment Cost vs. Annual Profit')
plt.xlabel('Investment Cost')
plt.ylabel('Annual Profit')
```

```python
inv_cost = [100, 200, 300, 400, 500]
ann_prof = [200, 400, 600, 800, 1000]

# 추세선이 있는 산점도 그리기
plt.scatter(inv_cost, ann_prof, color='skyblue')
m, b = np.polyfit(inv_cost, ann_prof, 1)
plt.plot(inv_cost, m*np.array(inv_cost) + b, color='red')
plt.title('Investment Cost vs. Annual Profit')
plt.xlabel('Investment Cost')
plt.ylabel('Annual Profit')
plt.show()

corr = np.corrcoef(inv_cost, ann_prof)[0,1]

if corr > 0.7:
  print(f"연간 수익과 투자 비용 간의 관계가 강합니다. 상관계수: {corr}")
else:
  print(f"연간 수익과 투자 비용 간의 관계가 약합니다. 상관계수: {corr}")
```
