Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
import pandas as pd
import numpy as np

sr1 = pd.Series([1,2,np.nan,4])
sr2 = pd.Series([5,6,7,8])

a = sr1.add(sr2, fill_value=0)
s = sr1.sub(sr2, fill_value=0)
m = sr1.mul(sr2, fill_value=0)
d = sr1.div(sr2, fill_value=0)

pd.DataFrame([a, s, m, d])
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])

r1 = df['A'] + s1
r2 = df['B'] + s2
r3 = df['C'] + s3

df2 = pd.DataFrame([r1, r2, r3], index=['A', 'B', 'C'])

for i in df2.index:
  df2.loc[[i], '합계'] = df2.loc[i].sum()

df2
```
