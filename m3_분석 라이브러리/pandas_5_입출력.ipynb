```python
import numpy as np
import pandas as pd
```

```python
# csv 파일 읽고 저장하기

ar1 = np.arange(10).reshape(2,5)
df = pd.DataFrame(ar1,columns=list('abcde'))

# to_csv할 때 index=None을 주면 read_csv시 index_col을 주지 않아도 됨.
df.to_csv('test.csv')
df = pd.read_csv('test.csv', index_col=0)

df.to_csv('test.csv', index=None)
df = pd.read_csv('test.csv')
```

```python
df1 = df.copy()
df1.to_csv('test1.csv')
df1 = pd.read_csv('test1.csv', index_col=0)
df1
```

```python
df1.iloc[0] = 2
df1.iloc[1] = 3
df1
```

```python
# json 파일 읽고 쓰기
df1.to_json('test2.json')
df_j = pd.read_json('test2.json')
df_j
```

```python
# 바이너리 파일(이진파일) 읽고 쓰기
df1.to_pickle('test3.pkl')
df1 = pd.read_pickle('test3.pkl')
df1
```

```python
# Q. 간단한 데이터프레임 작성 후 csv, json, pickle 형태로 저장 후 다시 불러와서 출력하세요
data = {'value':np.random.randn(6)}

df = pd.DataFrame(data,
                  index=pd.period_range(start='2024-01-01', end='2024-06-04', freq='M'))
df.to_csv('test.csv')
df.to_json('test.json')
df.to_pickle('test.pkl')

cs = pd.read_csv('test.csv')
js = pd.read_json('test.json')
pk = pd.read_pickle('test.pkl')
print(cs,'\n')
print(js,'\n')
print(pk,'\n')
```
