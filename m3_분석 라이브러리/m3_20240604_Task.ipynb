Task6. 0일부터 10일까지의 시간 범위를 12시간 간격으로 생성하고, 각 간격에 대해 누적 시간을 계산하여 데이터프레임을 작성하세요.

```python
# 시간 범위 생성
timedelta_range = pd.timedelta_range(start=0, end='10D', freq='12H')

# 데이터프레임 생성
df = pd.DataFrame({
  'timedelta':timedelta_range,
  # 누적 시간을 시간 단위로 계산
  'cumulative_time':[timedelta.total_seconds()/3600 for timedelta in timedelta_range]
})

# timedelta에는 누적 시간을 계산해주는 total_seconds() 함수가 존재한다.
for t in timedelta_range:
  print(t.total_seconds())
```

Task8. 주어진 df에서 월별로 데이터의 합계를 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-15', '2022-02-01', '2022-02-15', '2022-03-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

# 월별 데이터 집계
df['month'] = df['date'].dt.to_period('M') # freq='M'으로 period타입 반환
monthly_sum = df.groupby('month')['value'].sum().reset_index()
print(monthly_sum)

df.info() # period[M]
```

Task9. 주어진 데이터프레임에서 주말(토요일과 일요일) 데이터를 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-02', '2022-01-03', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])
df['dow'] = df['date'].dt.dayofweek # 월요일이 0 토,일이 5~6
weekend_of = df[df['dow'] >= 5]
weekend_of

#df['dow'] = df['date'].dt.day_name() # Saturday, Sunday
```

Task10. 주어진 df에서 분기별 데이터의 평균을 계산하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01', '2022-05-01'],
    'value': [1, 2, 3, 4, 5]
})
df['date'] = pd.to_datetime(df['date'])

df['quarter'] = df['date'].dt.to_period('Q')
quarterly_mean = df.groupby('quarter')['value'].mean().reset_index()
quarterly_mean
```

Task11. 주어진 df에서 작업 일자(월요일부터 금요일)만 남기고 필터링하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    'date': ['2022-01-01', '2022-01-03', '2022-01-04', '2022-01-05', '2022-01-08', '2022-01-09'],
    'value': [1, 2, 3, 4, 5, 6]
})
df['date'] = pd.to_datetime(df['date'])
df['dow'] = df['date'].dt.dayofweek
workdays_df = df[df['dow'] < 5]
workdays_df
```
