```python
# 데이터 프레임으로 불러오기
import cx_Oracle
import pandas as pd

# 데이터베이스 연결
con = cx_Oracle.connect("c##hr", "hr", "localhost:1521/xe")

#데이터베이스와의 상호작용을 위한 커서를 생성.
cur = con.cursor()

# EMPLOYEES 테이블의 모든 데이터를 선택하는 쿼리를 실행.
cur.execute('select * from employees')
# 쿼리 결과를 모두가져와 res 변수에 저장.
res = cur.fetchall()
# 데이터를 데이터프레임으로 변환
df = pd.DataFrame(res)
df
```

```python
# employees 테이블의 컬럼 이름을 조회하는 쿼리를 실행
cur.execute("select column_name from user_tab_columns where table_name='EMPLOYEES' order by column_id") # 테이블명 대문자
# 컬럼 이름 결과를 모두 가져와 col 변수에 저장
col = cur.fetchall()
# 컬럼 이름 리스트 생성
li = []
for i in col:
    for j in i:
        li.append(j)
# 데이터프레임 컬럼 이름 설정
df.columns = li
print(df.head())
# 커서를 닫는다.
cur.close()
# 데이터베이스 연결을 닫는다.
con.close()
```

```python
# drop
import cx_Oracle
conn = cx_Oracle.connect("c##hr", "hr", "localhost:1521/xe")
cs = conn.cursor()
sql = "drop table sample1"
cs.execute(sql)
cs.close()
conn.commit()
conn.close()
```

```python
# create
import cx_Oracle
conn = cx_Oracle.connect("c##hr", "hr", "localhost:1521/xe")
cs = conn.cursor()
sql = "create table sample1(id varchar2(10), score number)"
cs.execute(sql)
```

```python
# insert
sql = "insert into sample1 (id, score) values(1, 100)"
cs.execute(sql)
# conn.commit()
```

```python
# insert(여러개 입력)
# :1과 :2는 바인딩 변수로, 나중에 값을 대체하여 사용
for i in range(10):
    sql = "insert into sample1 values(:1, :2)"
    cs.execute(sql, (i, 100*i))
conn.commit()
```

```python
cs.execute('select * from sample1')
res = cs.fetchall()
df = pd.DataFrame(res)
df
```

```python
# update
sql = "update sample1 set score=50 where id='0'"
cs.execute(sql)
cs.close()
conn.commit()
conn.close()
```

#### 바인드 변수
- 인덱스를 사용하여 데이터 시퀀스의 각 요소를 SQL 문에 바인딩

#### 바인드 변수의 장점:
- 보안: SQL 인젝션 공격을 방지할 수 있다. 바인드 변수를 사용하면 입력 데이터가 SQL 구문으로 해석되지 않고 단순한 값으로 처리된다.- 
- 성능: 동일한 SQL 명령어를 반복 실행할 때 SQL 문을 파싱하고 컴파일하는 시간을 절약할 수 있다
- 가독성: 코드를 보다 명확하게 하고 유지보수를 쉽게 할 있다.
