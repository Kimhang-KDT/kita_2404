Task1_0528. 연산메소드 이용 및 NaN 값을 0으로 대체 후 사칙연산을 수행한 후 결과를 출력하세요.

```python
import pandas as pd
import numpy as np

sr1 = pd.Series([1,2,np.nan,4])
sr2 = pd.Series([5,6,7,8])

a = sr1.add(sr2, fill_value=0)
s = sr1.sub(sr2, fill_value=0)
m = sr1.mul(sr2, fill_value=0)
d = sr1.div(sr2, fill_value=0)

pd.DataFrame([a, s, m, d])
```

Task3_0528. 주어진 DataFrame의 여러 열에 대해 각기 다른 Series를 더하고, 결과를 새로운 DataFrame으로 반환한 후 각 행의 합계를 계산하여 새로운 열에 추가하여 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [1, 2, 3, 4, 5],
    "B": [10, 20, 30, 40, 50],
    "C": [100, 200, 300, 400, 500]
})

# Series 생성
s1 = pd.Series([5, 5, 5, 5, 5])
s2 = pd.Series([10, 10, 10, 10, 10])
s3 = pd.Series([15, 15, 15, 15, 15])

df_new = df.copy()

# 여러 열에 Series를 더하고 새로운 DataFrame에 저장
df_new['A'] = df['A'] + s1
df_new['B'] = df['B'] + s2
df_new['C'] = df['C'] + s3

# 각 행을 더하여 새로운 컬럼에 값을 추가
df_new['Sum'] = df_new.sum(axis=1)
df_new
```

Task4_0528. df에서 세 열의 값을 더하여 새로운 컬럼을 생성한 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [10, 20, 30, 40, 50],
    "B": [1, 2, 3, 4, 5],
    "C": [5, 10, 15, 20, 25],
    "D": [3, 6, 9, 12, 15],
    "E": [2, 4, 6, 8, 10]
})

# A와 B와 C컬럼 값을 행마다 각각 더해서 새로운 컬럼에 추가
for i in df.index:
  df.loc[i, 'SUM'] = df.loc[i,'A':'C'].sum()

# for문을 쓰지 않아도 바로 더하는 것이 가능하다
df['Sum'] = df['A'] + df['B'] + df['C']
df
```

Task5_0528. df에서 세 열의 값을 평균하여 새로운 컬럼을 생성 후 출력하세요.

```python
import pandas as pd

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [100, 200, 300, 400, 500],
    "B": [10, 20, 30, 40, 50],
    "C": [1, 2, 3, 4, 5],
    "D": [7, 8, 9, 10, 11],
    "E": [3, 6, 9, 12, 15]
})

# A부터 C까지의 값의 평균을 새로운 컬럼에 추가
for i in df.index:
  df.loc[i, 'MEAN'] = df.loc[i, 'A':'C'].mean()
df

# for문 사용X
df['MEAN'] = df[['A','B','C']].mean(axis=1)
df['MEAN'] = df.loc[:,'A':'C'].mean(axis=1)

df
```

Task6_0528. df에서 두 개의 열을 더한 값이 다른 한 개의 열보다 큰 경우에는 1, 작은 경우에는 0으로 값을 정하는 새로운 열을 생성한 후 출력하세요.

```python
import pandas as pd
import numpy as np

# 데이터프레임 생성
df = pd.DataFrame({
    "A": [2, 3, 4, 5, 6],
    "B": [1, 2, 3, 4, 5],
    "C": [3, 4, 5, 6, 7],
    "D": [8, 9, 10, 11, 12],
    "E": [10, 20, 30, 40, 50]
})

for i in df.index:
  cp = df.loc[i, 'A':'B'].sum()

  df.loc[i, 'BOOL'] = np.where(cp > df.loc[i, 'C'], 1, 0)

print(df)

##################
df['THAN'] = (df['A']+df['B'] > df['C']).astype(int)
df['THAN'] = (df['A']+df['B'] > df['C']).astype(bool)

df
```
