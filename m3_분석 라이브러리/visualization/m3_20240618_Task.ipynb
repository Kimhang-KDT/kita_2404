Task1_0618. auto-mpg 데이터 셋에서 mpg와 weight의 선형 관계를 그래프로 작성하세요.(전체, origin에 따른 관계 2개 작성)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('drive/MyDrive/KDT_2404/m3_분석 라이브러리/dataset/auto-mpg.csv')
```

```python
# origin에 따른 mpg와 weight의 선형 관계 그래프
plt.figure(figsize=(6,4))
sns.lmplot(data=df, x='weight', y='mpg', hue='origin')
plt.title('sns.Implot Example')
plt.show()
```

Task2_0618. pairplot을 생성하여 penguins 컬럼 간의 관계를 시각화 하세요. 단, 종(species)에 따라 다른 색상을 사용

```python
df = sns.load_dataset('penguins')
#df.species.unique()
# bill_length_mm, bill_depth_mm, flipper_length_mm, body_mass_g

sns.pairplot(df[df.columns], palette='rainbow',
             hue='species', diag_kind='kde')
```

Task3_0618. 펭귄 데이터셋을 사용하여, 각 종별로 섬(island)별 개체 수를 나타내는 카운트 플롯을 생성하세요. x축은 'island', hue는 'species'로 설정하세요.

```python
plt.figure(figsize=(8,4))
sns.set_style('whitegrid')
sns.countplot(x='island', data=df, hue='species', palette='bone')
plt.title('Penguins Count Plot')
plt.show()
```

Task4_0618.sex와 class간의 생존율을 보여주는 히트맵을 작성하세요.

```python
df = sns.load_dataset('titanic')

df['sex'].replace({'male':0, 'female':1}, inplace=True)

sns.heatmap(df[['survived', 'pclass', 'sex']].corr(), annot=True, cmap='Pastel1')
plt.title('Heatmap of the Dataset', pad=20, fontsize=20)
```

Task5_0618. 타이타닉 데이터셋을 사용하여, 객실 등급(class)에 따른 탑승객의 나이를 나타내는 박스 플롯과 바이올린 플롯을 같은 그래프 내에서 나란히 생성하세요. 이를 통해 각 클래스별 나이 분포의 차이를 비교하세요.

```python
# pclass에 따른 탑승객의 나이를 나타내는 박스플롯과 바이올린 플롯

sns.set_style('whitegrid')
sns.boxplot(x='class', y='age', data=df, color='skyblue')
sns.violinplot(x='class', y='age', data=df, color='lightgreen')
plt.title('Age by Class', pad=20, fontsize=20)
plt.xlabel('Age', labelpad=20, fontsize=15)
plt.ylabel('Class', labelpad=20, fontsize=15)
```

Task6_0618. 주어진 HTML 문서에서 모든 $<a>$ 태그를 추출하세요.

```python
from bs4 import BeautifulSoup

html_doc = """
<html><head><title>The Dormouse's story</title></head>
<body>
<p class="title"><b>The Dormouse's story</b></p>
<a href="http://example.com/elsie" class="sister" id="link1">Elsie</a>
<a href="http://example.com/lacie" class="sister" id="link2">Lacie</a>
<a href="http://example.com/tillie" class="sister" id="link3">Tillie</a>
</body></html>
"""

res = BeautifulSoup(html_doc, 'html.parser')
```

```python
res.find_all('a')
```
